```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from pandas.api.types import is_string_dtype, is_object_dtype
from rfpimp_MC import *
```

```python
df = pd.read_csv('../data collection/Pokemon.csv')
trim_df=df
targetVal='Type 1'

forbidden=['#', 'name', 'generation']

for entry in forbidden:
    trim=[]
    for f in trim_df.columns:
        if entry in f.lower() and not f==targetVal:
            trim.append(f)
    trim_df=trim_df.drop(trim, axis=1)

basefeatures = []

for f in trim_df.columns:
    if f!=targetVal:
        basefeatures.append(f)

def df_string_to_cat(trim_df):
    for col in trim_df.columns:
        if is_object_dtype(trim_df[col]) or is_string_dtype(trim_df[col]):
            trim_df[col] = trim_df[col].astype('category').cat.as_ordered()

def df_cat_to_catcode(trim_df):
    for col in trim_df.columns:
        if isinstance(trim_df[col].dtype, pd.CategoricalDtype):
            trim_df[col] = trim_df[col].cat.codes + 1

df_string_to_cat(trim_df)
df_cat_to_catcode(trim_df)

def fix_missing_num(trim_df, colname):
    trim_df[colname+'_na'] = pd.isnull(trim_df[colname])
    trim_df[colname].fillna(trim_df[colname].median(), inplace=True)

for f in trim_df.columns:
    fix_missing_num(trim_df, f)

X = trim_df[basefeatures]
y = trim_df[targetVal]

X_df = pd.DataFrame(X, columns=trim_df[basefeatures])
y_df = pd.DataFrame(y, columns=[targetVal])

X_df_train, X_df_test, y_df_train, y_df_test = train_test_split(X_df, y_df, test_size=0.2, random_state=123)

rf_classify = RandomForestClassifier(n_estimators=150, n_jobs=-1, oob_score=True)
rf_classify.fit(X, y)

new_data = pd.DataFrame({'Type 2': 0, 'Total': [14], 'HP':[318], 'Attack':[45],'Defense':[49],'Sp. Atk':[49],'Sp. Def':[65], 'Speed':[65], 'Legendary':False})
print(rf_classify.predict(pd.DataFrame(new_data, index=[0])))
```
